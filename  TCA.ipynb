```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
data=pd.read_csv('Customer Churn.csv')
data.head()
```

```python
data.info()
```

```python
data['TotalCharges']=data['TotalCharges'].replace(' ','0')
data['TotalCharges']=data['TotalCharges'].astype('float')
```

```python
data.isnull().sum().sum()
```

```python
data.describe()
```

```python
data['customerID'].duplicated().sum()
```

```python
data.head()
```

```python
def con(value):
    if value==1:
        return 'Yes'
    else:
        return "No"
data['SeniorCitizen']=data['SeniorCitizen'].map(con)
```

```python
ax=sns.countplot(x='Churn',data=data)
ax.bar_label(ax.containers[0])
plt.title('Count Of Customer Churn')
plt.show()
```

```python
plt.figure(figsize=(4,3))
data['Churn'].value_counts().plot(kind='pie',autopct="%1.2f%%")
plt.title('Percentage Of Churn Customer',fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='gender',data=data,hue='Churn')
plt.title('Churn By Gender')
plt.show()
             
```

```python
data.head()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='SeniorCitizen',data=data,hue='Churn')
plt.title('Churn By SeniorCitizen')
plt.show()
             
```

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(x='SeniorCitizen',data=data)
ax.bar_label(ax.containers[0])
plt.title('Count of Customer of SeniorCitizen')
plt.show()
```

```python
ct = pd.crosstab(
    data['SeniorCitizen'],
    data['Churn'],
    normalize='index'
) * 100

ax = ct.plot(kind='bar', stacked=True, figsize=(5,4))

# Add percentage labels
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', label_type='center')

plt.title('Churn By SeniorCitizen (%)')
plt.xlabel('SeniorCitizen')
plt.ylabel('Percentage')
plt.legend(title='Churn')
plt.show()
```

```python
plt.figure(figsize=(13,6))
sns.histplot(x='tenure',data=data,bins=72,hue='Churn')
plt.title('Histplot of Churn by Tenure')
plt.show()
```

```python
plt.figure(figsize=(4,4))
ax=sns.countplot(x='Contract',data=data,hue='Churn')
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title('Count of Customer of Contract')
plt.show()
```

```python
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

fig, axes = plt.subplots(3, 3, figsize=(18, 15))
axes = axes.flatten()

for i, col in enumerate(cols):
    sns.countplot(
        data=data,
        x=col,
        hue='Churn',
        ax=axes[i],
        palette='mako'
    )

    axes[i].set_title(col)
    axes[i].tick_params(axis='x', rotation=25)

    # Add count labels
    for container in axes[i].containers:
        axes[i].bar_label(container, fontsize=8)

plt.tight_layout()
plt.show()
```
